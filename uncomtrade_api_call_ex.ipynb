### API 등록키를 가지고 있는 경우
- 등록된 키를 가지고 있는 경우는 많은 양의 데이터를 한꺼번에 추출 가능

```python
mydf2 = comtradeapicall.getFinalData(subscription_key,  # Uncomptrade API 등록키
                                    typeCode='C',      # C: product, S:service
                                    freqCode='A',      # A: annual, M: monthly
                                    clCode='HS',
                                    period='2023,2024',     # 추출대상 년 혹은 월
                                    reporterCode=None,
                                    cmdCode='846021',   # HScode 2, 4, 6 자리로 설정 -> ex 개와 고양이 사료 HS: 230910
                                    flowCode='m',     # X: export, M:import
                                    partnerCode=None,
                                    partner2Code=None,
                                    customsCode=None, motCode=None, maxRecords=15000, format_output='JSON',
                                    aggregateBy=None, breakdownMode='classic', countOnly=None, includeDesc=True)
mydf2
```

```python
# mydf2 데이터셋에서 cmdCode와 customCode 컬럼을 맨앞으로 해서 display 해주세요.

# Get the list of columns
cols = mydf2.columns.tolist()

# Move 'cmdCode' and 'customsCode' to the front
cols.insert(0, cols.pop(cols.index('customsCode')))
cols.insert(0, cols.pop(cols.index('cmdCode')))

# Create a new DataFrame with the reordered columns
mydf_reordered = mydf2[cols]

# Display the reordered DataFrame
display(mydf_reordered)
```
